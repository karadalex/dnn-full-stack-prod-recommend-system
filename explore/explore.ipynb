```python
import pandas as pd

df1 = pd.read_csv("../datasets/ml-latest-small/movies.csv")
df1.head()
```

```python
df2 = pd.read_csv("../datasets/ml-latest-small/links.csv")
df2.head()
```

```python
df3 = pd.read_csv("../datasets/movie-genres-posters/MovieGenre.csv")
df3.head()
```

```python
df = df1.merge(df2, left_on="movieId", right_on="movieId")
df.head()
```

```python
df = df.merge(df3, left_on="imdbId", right_on="imdbId")
df.head()
```

```python
from IPython.display import Image

Image(url=df["Poster"][0])
```

```python
import json

# export csv with images
json_str = df.to_json(orient="records")
parsed = json.loads(json_str)

with open("../datasets/merged-small.json", 'w') as json_file:
  json_file.write(json.dumps({"data": parsed}, indent=4 ))
```
