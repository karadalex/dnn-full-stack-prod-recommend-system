```python
import pandas as pd

df1 = pd.read_csv("../datasets/ml-latest-small/movies.csv")
df1.head()
```

```python
df2 = pd.read_csv("../datasets/ml-latest-small/links.csv")
df2.head()
```

```python
df3 = pd.read_csv("../datasets/movie-genres-posters/MovieGenre.csv")
df3.head()
```

```python
df = df1.merge(df2, left_on="movieId", right_on="movieId")
df.head()
```

```python
df = df.merge(df3, left_on="imdbId", right_on="imdbId")
df.head()
```

```python
from IPython.display import Image

Image(url=df["Poster"][0])
```

```python
import json

# export csv with images
json_str = df.to_json(orient="records")
parsed = json.loads(json_str)

with open("../datasets/merged-small.json", 'w') as json_file:
  json_file.write(json.dumps({"data": parsed}, indent=4 ))
```

```python
df4 = pd.read_csv("../datasets/ml-latest-small/ratings.csv")
df4.head()
```

```python
def get_genres_list(genres):
  return genres.split("|")

df1["genres"] = df1["genres"].map(get_genres_list)
```

```python
for row in df1["genres"]:
  for genre in row:
    if genre not in df1.columns:
      df1[genre] = 0

df1.head()
```

```python
for i in df1.index:
  genres = df1["genres"][i]
  for genre in genres:
    df1[genre][i] = 1

df1.head()
```

```python
df5 = df1.merge(df4, left_on="movieId", right_on="movieId")
# Clean dataframe from unecessary columns
del df5["title"]
del df5["genres"]
del df5["timestamp"]
del df5["(no genres listed)"]
df5 = df5.loc[:, ~df5.columns.str.contains('^Unnamed')]
df5.head()
```

```python
df5.to_csv("../datasets/prepared-dataset-small.csv")
```
